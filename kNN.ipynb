```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import roc_auc_score, roc_curve, auc
```

Перекодируем значения целевой переменной в цифровые значения

```python
y = y.apply(lambda x: 1 if x=='M' else 0)
y
```

```python
y.value_counts(normalize=True)
```

Выполняем стандартизацию данных (в данном случае используем Минимакс), это необходимо т.к. значения разных парамметров имеют разные диапазоны значений и могут вносить разный вклад в предсказание.

```python
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)
x_1_scaled = scaler.fit_transform(x_1)
```

Разделяем дата сет на тренеровочный и тестовый

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.3, random_state=42, stratify=y) #train-test из полного датасета
x1_train, x1_test, y1_train, y1_test = train_test_split(x_1_scaled, y, test_size=0.3, random_state=42, stratify=y) #train-test из датасета c убранными кореллируемыми парамметрами
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
x1_train.shape, x1_test.shape, y1_train.shape, y1_test.shape
```

```python
y_train.value_counts(normalize=True)
```

```python
y_test.value_counts(normalize=True)
```

Строим модель kNN "из коробки" без настройки параметров для всего датасета. Высчитываем метрики качества.

```python
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
```

```python
print(accuracy_score(y_pred, y_test))
print(precision_score(y_pred, y_test))
print(recall_score(y_pred, y_test))
print(f1_score(y_pred, y_test))
print(roc_auc_score(y_test, y_pred))
```

Строим модель kNN "из коробки" без настройки параметров для датасета с исключенными скоррелированными данными. Высчитываем метрики качества.

```python
knn = KNeighborsClassifier()
knn.fit(x1_train, y1_train)
y1_pred = knn.predict(x1_test)
```

```python
print(accuracy_score(y1_pred, y1_test))
print(precision_score(y1_pred, y1_test))
print(recall_score(y1_pred, y1_test))
print(f1_score(y1_pred, y1_test))
print(roc_auc_score(y1_test, y1_pred))
```

Можем сделать вывод, что уделение скорелированных параметров, немного ухудшило качество предсказания, но не на много при этом сократив количество первоначальных признаков с 30 до 18.

```python
# Вычисляем ROC-кривую
fpr, tpr, thresholds = roc_curve(y_test, y_pred)

# Вычисляем AUC
roc_auc = auc(fpr, tpr)

# Строим график
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

Проведем подбор количества соседий методом перебора для accuracy и f1_score.

```python
neighbors = range(1, 50)
accuracy_train = []
accuracy_test = []
for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    preds = knn.predict(x_train)
    preds_test = knn.predict(x_test)
    accuracy_train.append(accuracy_score(preds, y_train))
    accuracy_test.append(accuracy_score(preds_test, y_test))
    
plt.plot(neighbors, accuracy_train, color='blue', label='train')
plt.plot(neighbors, accuracy_test, color='red', label='test')
plt.title("Max test quality: {:.3f}\nBest k: {}".format(max(accuracy_test), np.argmax(accuracy_test)+1))
plt.legend()
plt.show()
```

```python
neighbors = range(1, 50)
f1_score_train = []
f1_score_test = []
for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    f1_score_train.append(f1_score(knn.predict(x_train), y_train))
    f1_score_test.append(f1_score(knn.predict(x_test), y_test))
    
plt.plot(neighbors, f1_score_train, color='blue', label='train')
plt.plot(neighbors, f1_score_test, color='red', label='test')
plt.title("Max test quality: {:.3f}\nBest k: {}".format(max(f1_score_test), np.argmax(f1_score_test)+1))
plt.legend();
```

Определим оптимальное количество соседий с помощью кросс-валидации на всем датасете, разбивая его на 5 частей, и высчитывая среднюю по accuracy.

```python
from sklearn.model_selection import cross_val_score
neighbors = range(1, 50)
mean_cross_val = []
for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_scaled, y)
    mean_cross_val.append((cross_val_score(knn, x_scaled, y, cv=5, scoring='accuracy')).mean())
plt.plot(neighbors, mean_cross_val, color='blue', label='mean_cross_val')
plt.title("Max mean_cross_val: {:.3f}\nBest k: {}".format(max(mean_cross_val), np.argmax(mean_cross_val)+1))
plt.legend();

```

В соответствие с кросс-валидацией оптимальное количество соседий 6, хотя результат практически одинаков с 9 и 10 соседями. 

Пересчетаем метрика качества для тестовой выборки для 9 соседей.

```python
knn = KNeighborsClassifier(n_neighbors=9)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
print(accuracy_score(y_pred, y_test))
print(precision_score(y_pred, y_test))
print(recall_score(y_pred, y_test))
print(f1_score(y_pred, y_test))
print(roc_auc_score(y_test, y_pred))
```
